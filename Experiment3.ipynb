```python
# Extract first 500 queries from developer data (with existing qrel data)
import csv

with open('collections/msmarco-passage/queries.dev.matched.tsv','r') as tsvIn, open('collections/msmarco-passage/queries.dev.500.tsv', 'w') as tsvOut:
    reader = csv.reader(tsvIn, delimiter='\t')
    writer = csv.writer(tsvOut, delimiter='\t')
    i=1;
    for row in reader:
        if (i<=500):
            writer.writerow(row)     
        i+=1
```
